```python
import os
import pandas as pd
import seaborn as sns
from dotenv import load_dotenv

# load the .env file variables
load_dotenv()

import spotipy
import matplotlib.pyplot as plt

from spotipy.oauth2 import SpotifyClientCredentials

# load the .env file variables
load_dotenv()
```

```python
#Insert the info to work with it
client_id = os.environ.get("CLIENT_ID")
client_secret = os.environ.get("CLIENT_SECRET")

auth_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
spotify = spotipy.Spotify(auth_manager=auth_manager)

breaking_benjamin_uri = '5BtHciL0e0zOP7prIHn3pP'
results = spotify.artist_top_tracks('5BtHciL0e0zOP7prIHn3pP')
print(results)
```

```python
#Since it returned a dictionary, lets confirm the Key
results.keys()
```

```python
df = pd.DataFrame(results['tracks'])
df.head()
```

```python
#Func to Convert milliseconds to minutes
def mil_convert(milliseconds):
   seconds, milliseconds = divmod(milliseconds, 1000)
   minutes, seconds = divmod(seconds, 60)
   return minutes, seconds
```

```python
# Pull the information using 'track' with the most played songs of the artist, keep the name of the song, the popularity and the duration (in minutes).
for track in results['tracks']:
    minutes, seconds = mil_convert(track['duration_ms'])
    print(f"Track         : {track['name']}")
    print(f"Popularity    : {track['popularity']}")
    print(f"Duration      : {minutes}:{seconds:02d} minutes")
    print()
```

```python
track_list = []

#ReBuild dictionary for DataFrame
for track in results['tracks']:
    track_info = {
        'name': track['name'],
        'popularity': track['popularity'],
        'duration_ms': track['duration_ms'],
        'duration_min': round(track['duration_ms'] / 1000 / 60, 2),
        'album_name': track['album']['name'],
        'release_date': track['album']['release_date'],
        'artist_name': track['artists'][0]['name']
        }
    track_list.append(track_info)

df = pd.DataFrame(track_list)
df
```

```python
#Sort by popularity
df_sorted = df.sort_values(by='popularity', ascending=False)

#Show top 3 songs
top_3_songs = df_sorted.head(3)
top_3_songs
```

## Song Duration vs Popularity

```python
#Visualize scatter plot

sns.scatterplot(data=df, x='duration_ms', y='popularity')
plt.title('Song Duration vs Popularity')
plt.ylabel('Popularity')
plt.xlabel('Duration (milliseconds)')
plt.show()
```

```python
correlation = df['duration_ms'].corr(df['popularity'])

if abs(correlation) >= 0.7:
    interpretation = "a strong relationship"
elif abs(correlation) >= 0.3:
    interpretation = "a moderate relationship"
elif abs(correlation) >= 0.1:
    interpretation = "a weak relationship"
else:
    interpretation = "no meaningful relationship"

direction = "positive" if correlation > 0 else "negative"

print(f"Correlation between duration and popularity: {correlation:.2f}")
print(f"This indicates {interpretation} ({direction}).")
```

## Conclusion

1. The correlation between duration and popularity is -0.11.
2. This indicates a weak relationship (negative).
3. The scatter plot shows a slight trend; songs of varying lengths can be equally popular. There is only a *weak* tendency, not enough to claim shorter songs are more popular.
4. **Conclusion**: Song duration does not impact popularity.
